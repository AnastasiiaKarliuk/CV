```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

from itertools import chain
```

# 

```python
img_path = "with_all_our_love/hearts 1.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img_list = list(chain.from_iterable(img))

plt.hist(img_list, bins=100)
plt.show()
```

```python
(T, img_bin) = cv2.threshold(img, 15, 255, cv2.THRESH_BINARY)
img_new = cv2.erode(img_bin, np.ones((5, 5), np.uint8), iterations=1)
plt.imshow(img_new, cmap='gray')
```

```python
img_path = "with_all_our_love/hearts 2.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img_bin = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 255, 2)
img_new = cv2.erode(img_bin, np.ones((5, 5), np.uint8), iterations=1)
plt.imshow(img_new, cmap='gray')
```

```python
img_path = "with_all_our_love/hearts 3.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
plt.hist(img[:, :].ravel(), bins=100)
plt.show()
```

```python
img_bin = np.where((img<80) | ((110<img) & (img<145)), 0, 255)
new_img = np.uint8(img_bin)

mask0 = np.zeros((9, 9), np.uint8)
img_new = cv2.morphologyEx(new_img, cv2.MORPH_CLOSE, mask0, 1)
plt.imshow(new_img, cmap='gray')
```

```python
img_path = "with_all_our_love/hearts 4.png"
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
rgb_img.shape
```

```python
r, g, b = cv2.split(rgb_img)
img_bin = np.where(((r<135) & (g>111)), 0, 255)
img_bin = np.uint8(img_bin)
```

```python
mask0 = np.zeros((9, 9), np.uint8)
img_new = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, mask0, 1)
plt.imshow(img_new, cmap='gray')
```

```python
img_path = "with_all_our_love/hearts 5.png"
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
plt.hist(img.ravel(), bins=100)
plt.show()
```

```python
img = cv2.medianBlur(img, 3)
plt.imshow(img, cmap='gray')
```

```python
plt.hist(img.ravel(), bins=100)
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=11, tileGridSize=(11,11))
s = clahe.apply(img)
plt.imshow(s, cmap='gray')
```

```python
img_bin = cv2.adaptiveThreshold(s, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 255, 3)
plt.imshow(img_bin, cmap='gray')
```
